# Plotting data using Pandas
As a beginner student, learning how to plot data using Pandas is a fundamental skill in data analysis. Pandas provides a wide range of functions to create different types of plots, allowing us to visualize data and gain insights into patterns, trends, and relationships. Let's explore the process of plotting data using Pandas:

1. **Importing the Required Libraries:**

 Before we start plotting, we need to import the necessary libraries: Pandas and Matplotlib. Matplotlib is a popular data visualization library that works seamlessly with Pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

2. **Loading the Data:**

 We begin by loading our data into a Pandas DataFrame. This can be done by reading data from a file (e.g., CSV, Excel) or by creating a DataFrame from existing data structures.

```python
df = pd.read_csv('data.csv')  # Replace 'data.csv' with the actual file path or URL
```

3. **Exploring the Data**:

 It's crucial to have a basic understanding of the data before plotting. We can examine the DataFrame's structure, preview a few rows, and check the data types and summary statistics.

```python
print("\n display HEAD function \n")
df.head()     # Display the first few rows of the DataFrame

```

```python
print("\ndisplay info function \n")
df.info()    # Get information about the DataFrame
```

```python
print("\ndisplay Describe function \n")
df.describe()  # Summary statistics of numerical columns
```

4. **Creating Plots:**

 Pandas provides a plot() function that we can apply directly to a DataFrame or a specific column. We can specify the type of plot we want by setting the kind parameter. Here are a few examples:

**a. Line Plot:**

  Visualize trends over time or continuous variables using a line plot.

```python
df.plot(kind='line', x='Class', y='Percantege')
plt.show()
```

**b. Bar Plot:**
Compare categorical variables or analyze frequency distributions using a bar plot.

```python
df.plot(kind='bar', x='Class', y='Percantege')
plt.show()
```

**c. Histogram:**
Visualize the distribution of numerical data using a histogram.

```python
df['Class'].plot(kind='hist')
plt.show()
```

```python
df['Percantege'].plot(kind='hist')
plt.show()
```

**d. Scatter Plot:**
Examine the relationship between two continuous variables using a scatter plot.

```python
df.plot(kind='scatter', x='Class', y='Percantege')
plt.show()
```

These are just a few examples of the plots you can create using Pandas. There are many more types of plots available, such as box plots, area plots, pie charts, and heatmaps. Each plot type serves a specific purpose and can be customized further based on your requirements.

**Customizing Plots:**
Pandas allows us to customize our plots to improve their visual appearance and convey information more effectively. We can modify various aspects of the plot, such as colors, labels, titles, axes, legends, and gridlines. For example:

```python
df.plot(kind='line', x='Class', y='Percantege', color='red', linestyle='--', marker='o')
plt.title('Value Over Time')
plt.xlabel('Class')
plt.ylabel('Percantege')
plt.legend()
plt.grid(True)
plt.show()
```

Here, we modified the line color, style, and marker shape. We also added a title, labeled the x and y axes, included a legend, enabled gridlines, and displayed the plot.

# Subplots and multiple plots on the same figure
As a beginner student, understanding how to create subplots and place multiple plots on the same figure in Pandas is valuable for organizing and comparing different visualizations. Let's explore how to achieve this:

1. **Creating Subplots:**

 Subplots allow you to divide a single figure into multiple smaller plots. Pandas provides an interface to create subplots using the plt.subplots() function. The syntax is as follows:

```python
fig, axes = plt.subplots(nrows, ncols)
```

- **nrows**: The number of rows of subplots in the figure.
- **ncols**: The number of columns of subplots in the figure.
- **fig**: The figure object representing the entire figure.
- **axes**: A NumPy array or a list of axes objects representing each subplot.

For example, to create a figure with 2 rows and 2 columns of subplots:

```python
fig, axes = plt.subplots(2, 2)
```

**This will generate a 2x2 grid of subplots.**

2. **Plotting on Subplots:**
Once you have created the subplots, you can plot data on each individual subplot by accessing the corresponding axes objects. You can use the indexing notation **axes[row_index, col_index]** to specify the subplot where you want to plot the data.

For example, to plot a line chart on the top-left subplot:

```python
axes[0, 0].plot(x1, y1)
```

**Similarly, you can plot on other subplots by modifying the indices accordingly.**

3. **Adjusting Subplot Layout:**
Pandas provides methods to adjust the layout of subplots within a figure. Some commonly used methods include:

 - plt.tight_layout(): Automatically adjusts the subplots to prevent overlapping.
 - fig.subplots_adjust(): Manually adjusts the spacing between subplots.
4. **Multiple Plots on the Same Figure:**
You can also place multiple plots on the same figure without using subplots. To achieve this, you can create separate plot calls and customize their positions within the figure using the plt.subplots() function. Here's an example:

```python
fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.plot(x2, y2)
ax.scatter(x3, y3)
```

In this example, we create a single plot with multiple lines and a scatter plot.

5. **Customizing Subplots:**
Each subplot can be customized independently. You can modify the labels, titles, legends, colors, and other properties of each individual subplot using the corresponding axes objects. This allows you to create distinct visualizations within the same figure.

6. **Saving Subplots:**
You can save the entire figure with all the subplots as an image file using the plt.savefig() function. This allows you to export the complete visual representation for further use or sharing.

By utilizing subplots and placing multiple plots on the same figure, you can efficiently compare and analyze different aspects of your data. Whether you want to compare trends, distributions, or relationships, subplots offer a flexible way to organize and present your visualizations. Experiment with different configurations and plot types to create meaningful and insightful figures.

**To demonstrate subplots and multiple plots on the same figure using the given dataset, we can create different visualizations for each attribute and display them together in a single figure. Let's consider the following example:**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('data.csv')

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Study Hours vs Percentage
axes[0, 0].scatter(df['Study hrs'], df['Percantege'])
axes[0, 0].set_xlabel('Study Hours')
axes[0, 0].set_ylabel('Percentage')

# Plot 2: Sleeping Hours vs Percentage
axes[0, 1].scatter(df['Sleeping hrs'], df['Percantege'])
axes[0, 1].set_xlabel('Sleeping Hours')
axes[0, 1].set_ylabel('Percentage')

# Plot 3: Social Media Usage Hours vs Percentage
axes[1, 0].scatter(df['Social Media usage hrs'], df['Percantege'])
axes[1, 0].set_xlabel('Social Media Usage Hours')
axes[1, 0].set_ylabel('Percentage')

# Plot 4: Mobile Games Hours vs Percentage
axes[1, 1].scatter(df['Mobile Games hrs'], df['Percantege'])
axes[1, 1].set_xlabel('Mobile Games Hours')
axes[1, 1].set_ylabel('Percentage')

# Adjust subplot spacing
plt.tight_layout()

# Show the plot
plt.show()
```

In this example, we have created a 2x2 grid of subplots using **plt.subplots(2, 2).** Each subplot represents a scatter plot between one attribute and the percentage. We access each subplot using the indexing notation **axes[row_index, col_index]** and then plot the corresponding attribute against the percentage.

We set appropriate labels for the x-axis and y-axis of each subplot using **set_xlabel() and set_ylabel().** Adjusting the spacing between subplots is done with **plt.tight_layout()** to prevent overlapping.

By running this code, you will see a figure with four scatter plots, each displaying the relationship between an attribute and the percentage. This allows for easy comparison and analysis of how different attributes relate to the percentage.

Feel free to customize the plot properties, such as colors, markers, and titles, to suit your preferences and enhance the visual appearance of the subplots.

# Saving and exporting plots
Once you have created your desired plot using Pandas, you may want to save it as an image file for further use or sharing. Pandas provides a convenient way to save plots using the savefig() function. Let's explore how to save and export plots:

**Saving a Plot:**

To save a plot, you can use the **savefig()** function from the pyplot module. The general syntax is as follows:

```python
plt.savefig('filename.extension')
```

filename: The name you want to give to the saved file.
extension: The desired file format, such as PNG, JPEG, PDF, SVG, etc.
For example, to save a plot as a PNG image named "my_plot.png"

```python
plt.savefig('my_plot.png')
```

**The saved file will be stored in the current working directory.**

**Specifying the File Format:**

By default, savefig() determines the file format based on the extension you provide. However, you can explicitly specify the file format using the format parameter. For instance

```python
plt.savefig('my_plot', format='pdf')
```

In this example, the plot will be saved as a PDF file named "my_plot.pdf".

3. **Saving with Custom DPI:**

You can also control the resolution (dots per inch) of the saved plot using the dpi parameter. A higher DPI value results in a higher quality image but also a larger file size. The default DPI value is typically 100. For example:

```python
plt.savefig('my_plot.png', dpi=300)
```

4. **Saving Subplots:**
If you have multiple subplots and want to save them as a single image, ensure that you call savefig() on the figure object (fig) rather than the individual subplot objects. For example:

```python
fig, axes = plt.subplots(2, 2)
# ... plotting code ...

fig.savefig('subplots.png')
```

This saves all the subplots as a single PNG image named "subplots.png".

Remember to call savefig() after you have completed all the desired plotting and customizations. The function saves the current active figure, so make sure you have only one figure or select the specific figure you want to save if there are multiple figures in your code.

By using the savefig() function, you can easily save your plots in various file formats, ensuring that you can access and share them for presentations, reports, or further analysis.
